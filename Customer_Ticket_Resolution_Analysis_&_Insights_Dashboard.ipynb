DATA CLEANING


```python
#renaming the columns in snake_case
df= df.rename(columns={'Ticket ID':'ticket_id', 'Customer Name': 'customer_name','Customer Email':'customer_email','Customer Age':'customer_age','Customer Gender':'customer_gender','Product Purchased':'product_purchased','Date of Purchase':'date_of_purchase','Ticket Type':'ticket_type','Ticket Subject':'ticket_subject','Ticket Description': 'ticket_description', 'Ticket Status': 'ticket_status', 'Resolution':'resolution','Ticket Priority': 'ticket_priority', 'Ticket Channel': 'ticket_channel', 'First Response Time': 'first_response_time', 'Time to Resolution': 'time_to_resolution', 'Customer Satisfaction Rating': 'customer_satisfaction_rating'})
df
```

```python
df.columns
```

```python
#changing the date columns datatype to datetime
df['date_of_purchase'] = pd.to_datetime(df['date_of_purchase'])
df['first_response_time'] = pd.to_datetime(df['first_response_time'])
df['time_to_resolution'] = pd.to_datetime(df['time_to_resolution'])
```

```python
df.info()
```

```python
#replacing the Null values to unknown or 0
for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = df[col].fillna('unknown')
  elif df[col].dtype == 'datetime64[ns]':
    df[col] = df[col].fillna(pd.to_datetime('1900-01-01'))
  else:
    df[col] = df[col].fillna(0)
```

```python
df.info()
```

```python
df = df[df['first_response_time'] <= df['time_to_resolution']]
df
```

KPI CALCULATIONS


```python
df
```

```python
#calculating Resolution Time
df['resolution_time_hours'] = (df['time_to_resolution']- df['first_response_time']).dt.total_seconds()/ 3600
df
```

```python
df['Month']= df['date_of_purchase'].dt.to_period('M')
df
```

```python
df['Day'] = df['date_of_purchase'].dt.day_name()
df
```

DATA VISUALIZATION


```python
import matplotlib.pyplot as plt
df['ticket_type'].value_counts().plot(kind ='bar')
plt.title('Tickets by category')
plt.xlabel('ticket type')
plt.ylabel('Number of tickets')
plt.xticks(rotation= 45)
plt.show()
```

Technical issue tickets are high

```python
#monthly trend
monthly_trend = df.groupby('Month')['ticket_id'].count()
monthly_trend.plot(kind = 'line')
plt.title('Monthly Ticket Trend')
plt.xlabel('Month')
plt.ylabel('Number of tickets')
plt.show()
```

Mid April of 2020 and 2021 had the lowest ticket count.
October 2020 had highest tickets

```python
#priority distribution
df['ticket_priority'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('Ticket Priority Distribution')
plt.ylabel('')
plt.show()
```

The distribution among different priority tickets is almost equal.

```python
#Resolution Time by Channel
df.groupby('ticket_channel')['resolution_time_hours'].mean().plot(kind = 'bar')
plt.title('Resolution Time by Channel')
plt.xlabel('Channel')
plt.ylabel('Resolution Time (hours)')
plt.show()
```

Resolution provided through phone channel is quickest.
Resolution provided through email channel is slow.

```python
plt.scatter(df['resolution_time_hours'], df['customer_satisfaction_rating'])
plt.title("Resolution Time vs Satisfaction")
plt.xlabel("Resolution Time (Hours)")
plt.ylabel("Satisfaction Rating")
plt.show()
```
